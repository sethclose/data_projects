```python
def xml_tier (sub_dict, tag, indent, text):
    text += indent + '<' + tag + '>' + '\n'
    indent += "   "
    for key, value in sub_dict.items():
        start_tag = '<' + key + '>'
        end_tag = '</' + key + '>'
        if type(value) == str:
            item = indent + start_tag + value + end_tag
            text += item + '\n'
        else:
            text = xml_tier(value, key, indent, text)
    indent = indent[:-3]
    text += indent + '</' + tag + '>' + '\n'
    return text 

def json_to_xml (json_list):
    name = 'Instance'
    indent = ''
    text = ''
    for json_dict in json_list:
        text = xml_tier(json_dict, name, indent, text)
    return text

print(json_to_xml(json_list))
```

```python
def xml_tier (sub_dict, tag, indent):
    print(indent + '<' + tag + '>')
    indent += "   "
    for key, value in sub_dict.items():
        start_tag = '<' + key + '>'
        end_tag = '</' + key + '>'
        if type(value) == str:
            item = indent + start_tag + value + end_tag
            print(item)
        else:
            xml_tier(value, key, indent)
    indent = indent[:-3]
    print(indent  + '</' + tag + '>')

def json_to_xml (json_list):
    for json_dict in json_list:
        xml_tier(json_dict, 'Instance', '')

json_to_xml(json_list)
```
